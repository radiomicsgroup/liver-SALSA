## Run inference with SALSA-liver

#### 1. Generate a csv with the absolute paths to the scan

You just need one column called 'PATHS' and in each row the absolute path to each of the scans

There is a code for generating this type of csv in ```/src/utils/csv.py```

````python
import os
import pandas as pd

path_folder = '/path/to/folder'
pts_inference = sorted([ i for i in os.listdir(path_folder)])
print(len(pts_inference), *pts_inference, sep = ', ')

paths = []
for pt in pts_inference:
    scan = [i for i in os.listdir(os.path.join(path_folder, pt)) if '.nii.gz' in i ]
    paths.append(os.path.join(path_folder, pt, scan[0]))

df = pd.DataFrame(paths, columns = ['PATHS'])
df.to_csv(os.path.join('/path/to/csv.csv'), index = False)
````

#### 2. Run SALSA

I have updated the way we used to run SALSA. We no longer need 2 environments or 2 codes, now you just need to create an environment with ```environment.yml``` and then execute 1 single command.

SALSA generates a segmentation mask (in .nii.gz) from a scan (also in .nii/.nii.gz format) saved in the same path as your scan with a '_SALSA' extension at the end of the scan filename. Moreover, there are intermediate files that will be stored in a tmp folder of the same name (scan filename + '_SALSA') that will be later removed (unless specified). In this ```demo``` folder there are examples of the resulting file structure.


The code has 2 new optional features:
* ```--keep_intermediate_files```: False by default, it is in order to keep/delete all the intermediate files stored in the tmp folder generated (useful for debugging purposes)
* ```--recist```: True by default, it is in order to apply a new layer post-processing based on RECIST-criteria. If applied, the non-measurable disease will be stored as 2s in the generated mask.

````python
python run.py --input_csv demo/paths_demo.csv --keep_intermediate_files False --recist True
````

In the ```demo``` folder there are the results as well as the GT files (in case you want to see how good SALSA works)
